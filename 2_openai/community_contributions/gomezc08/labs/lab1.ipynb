## Import Libraries

```python
from dotenv import load_dotenv
from openai import OpenAI
from agents import Agent, Runner, trace
from IPython.display import Markdown, display
```

```python
load_dotenv(override=True)
```

## Agent Definitions

```python
science_agent = Agent(name="Science Agent", instructions="You answer science related questions")
math_agent = Agent(name="Math Agent", instructions="You answer math related questions")
english_agent = Agent(name="English Agent", instructions="You answer english/humanities related questions")
```

Let's test one of em out

```python
with trace("Math Problem"):
    result = await Runner.run(math_agent, "Find what x is here and provide steps: 2x+4 = 20")
```

```python
display(Markdown(result.final_output))
```
