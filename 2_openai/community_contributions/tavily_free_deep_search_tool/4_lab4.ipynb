````python
tavily_api_key = os.getenv('TAVILY_API_KEY')
if(tavily_api_key):
    print(f"Tavily API Key Exists: {tavily_api_key}")
else:
    print("No Tavily API Key")
    
tavily_client = TavilyClient(os.getenv('TAVILY_API_KEY'))

@function_tool
def tavily_search_tool(query: str):
    print(f"---Travily is searching for query")
    # can use search_depth="advanced" for advance search
    return tavily_client.search(query,  search_depth="basic")
````

````python
INSTRUCTIONS = "You are a research assistant. Given a search term, you search the web for that term and \
produce a concise summary of the results. The summary must 2-3 paragraphs and less than 300 \
words. Capture the main points. Write succintly, no need to have complete sentences or good \
grammar. This will be consumed by someone synthesizing a report, so it's vital you capture the \
essence and ignore any fluff. Do not include any additional commentary other than the summary itself."

# Instead of WebSearchTool(search_context_size="low") we will use tavily_search_tool

search_agent = Agent(
    name="Search agent",
    instructions=INSTRUCTIONS,
    tools=[tavily_search_tool],
    model="gpt-4o-mini",
    model_settings=ModelSettings(tool_choice="required"),
)

````

````python
message = "Latest AI Agent frameworks in 2025"

with trace("Deep Search"):
    result = await Runner.run(search_agent, message)

display(Markdown(result.final_output))
````
