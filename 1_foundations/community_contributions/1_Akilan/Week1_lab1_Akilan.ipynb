```python
from dotenv import load_dotenv
import os
from IPython.display import Markdown, display
from openai import OpenAI
load_dotenv(override=True)
openai_api_key=os.getenv("OPENAI_API_KEY")
openai = OpenAI(api_key=openai_api_key)
```

```python
question = "Pick a business area that might be worth exploring for an Agentic AI opportunity."
model_name = "gpt-4.1-nano"
messages = [{"role": "user", "content": question}]
response = openai.chat.completions.create(
    model=model_name,
    messages=messages)
result = response.choices[0].message.content
display(Markdown(result))
```

```python
question2 = "Present a pain-point in " + result + " something that might be ripe for an Agentic AI solution."
messages2 = [{"role": "user", "content": question2}]
response2 = openai.chat.completions.create(
    model=model_name,
    messages=messages2)
result2 = response2.choices[0].message.content
display(Markdown(result2))
```

```python
question3 = "Propose an Agentic AI solution for " + result2
messages3 = [{"role": "user", "content": question3}]
response3 = openai.chat.completions.create(
    model=model_name,
    messages=messages3)
result3 = response3.choices[0].message.content
display(Markdown(result3))
```
