### Import Lib


```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv(override=True)

openrouter_base_url = os.getenv('OPENROUTER_API_BASE_URL')
openrouter_api_key = os.getenv('OPENROUTER_API_KEY')

if openrouter_base_url:
    print(f"OpenRouter Base URL exists {openrouter_base_url}")
else:
    print("OpenRouter Base URL not set - please head to the troubleshooting guide in the setup folder")

if openrouter_api_key:
    print(f"OpenRouter API Key exists and begins {openrouter_api_key[:8]}")
else:
    print("OpenRouter API Key not set - please head to the troubleshooting guide in the setup folder")
```

### Setup OpenAI


```python
openrouter = OpenAI(
  base_url=openrouter_base_url,
  api_key=openrouter_api_key,
)
```

### Generate for Business Idea


```python
messages = [{"role": "user", "content": "Please pick a business area that might be worth exploring for an Agentic AI opportunity"}]

response = openrouter.chat.completions.create(
  model="openai/gpt-4.1-mini",
  messages=messages
)

business_idea = response.choices[0].message.content

business_idea
```

### Generate for pain point


```python
messages = [{"role": "user", "content": f"""Please present a pain point in this industry. 

{business_idea}
"""}]

response = openrouter.chat.completions.create(
  model="openai/gpt-4.1-mini",
  messages=messages
)

pain_point = response.choices[0].message.content

pain_point
```

### Generate for solution


```python

messages = [{"role": "user", "content": f"""Please propose the Agentic AI Solution for this Pain Point. 

{pain_point}
"""}]

response = openrouter.chat.completions.create(
  model="openai/gpt-4.1-mini",
  messages=messages
)

solution = response.choices[0].message.content

solution
```
