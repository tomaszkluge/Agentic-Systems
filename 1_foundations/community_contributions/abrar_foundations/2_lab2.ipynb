```python
print(judge)
```

## Another agentic design pattern: Reflection (Critique-and-Refine)

The notebook already used two patterns:
1. **Multi-model competition** – same question sent to several LLMs; each returns an answer.
2. **Judge** – one LLM evaluates all answers and ranks them (e.g. JSON with `results`).

Here we add a third pattern: **Reflection (Critique-and-Refine)**. We take the **winning** answer, have a **critic** LLM evaluate it (strengths, weaknesses, one concrete improvement), then a **refiner** LLM produces an improved synthesis. This improves quality by iterating on the best candidate instead of accepting it as-is.

```python
# Get the winning model's answer (rank 1 from the judge)
# ranks contains 1-based competitor numbers: e.g. ["2", "1", "3"] means competitor 2 is best
winner_one_based = int(ranks[0])
winner_idx = winner_one_based - 1
winning_model = competitors[winner_idx]
winning_answer = answers[winner_idx]

print(f"Winning model: {winning_model}")
print(f"Winning answer (first 300 chars):\n{winning_answer[:300]}...")
```

```python
# Critic agent: evaluate the winning answer (strengths, weaknesses, one concrete improvement)
critic_prompt = f"""You are a critical reviewer. The following question was posed to several LLMs, and one answer was judged best.

Question:
{question}

Winning answer:
{winning_answer}

Provide a short critique (2-4 sentences) with:
1. One strength of the answer.
2. One weakness or gap.
3. One specific, concrete improvement (what to add or change).

Be concise. No preamble."""

critic_messages = [{"role": "user", "content": critic_prompt}]
response = openai.chat.completions.create(model="gpt-4o-mini", messages=critic_messages)
critique = response.choices[0].message.content
print("Critique:")
print(critique)
```

```python
# Refiner agent: produce an improved synthesis using the original answer and the critique
refiner_prompt = f"""You are an editor. Given the original question, the best existing answer, and a critic's feedback, produce an improved version of the answer.

Question:
{question}

Best existing answer:
{winning_answer}

Critic's feedback:
{critique}

Instructions: Write an improved answer that keeps the strengths, addresses the weakness, and incorporates the suggested improvement. Stay similar in length and style. Output only the improved answer, no meta-commentary."""

refiner_messages = [{"role": "user", "content": refiner_prompt}]
response = openai.chat.completions.create(model="gpt-4o-mini", messages=refiner_messages)
refined_answer = response.choices[0].message.content

print("Refined (improved) answer:")
display(Markdown(refined_answer))
```
