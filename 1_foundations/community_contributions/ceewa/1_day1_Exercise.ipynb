```python
# First create the messages:
system_prompt = """You are a venture capitalist, enterprise consultant, and expert AI product strategist. 
    Your goal is to identify a single, high-value business area ripe for an Agentic AI opportunity.
    An ideal business area must feature operational workflows that require autonomous reasoning, 
    multi-step planning, and cross-platform execution—not just static content generation or basic data entry.
    Constraints:
        Industry Focus: Fintech
        Data Limits: Focus only on workflows that do not require highly specialized, proprietary data. 
                    The workflow must rely on standard data structures, open APIs, and common business software.
    Your task is to choose exactly one business area and provide your response in this precise format:
        Targeted Business Area: [The specific niche workflow or department]
        Core Purpose: [A clear, one-sentence description explaining exactly what this workflow achieves for a business]
"""

messages = [{"role": "user", "content": system_prompt}]

#Debug test to ensure variables aren't from your .env file
print("Loaded Base URL: ", os.getenv("GEMINI_BASE_URL"))

# Then make the first call:
client = OpenAI(
    api_key=os.getenv("GOOGLE_API_KEY"), 
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
    )

response = client.chat.completions.create(
    model="gemini-2.5-flash",
    messages=[{"role": "user", "content": system_prompt}]
)
print(response.choices[0].message.content)
business_response = response.choices[0].message.content

# Then read the business idea:

business_idea = business_response

system_prompt = f"""
    You are a senior operations auditor and enterprise workflow researcher. 
    Your task is to analyze a specific business area and identify its most painful, 
    costly operational bottleneck—something highly challenging for teams to manage 
    but perfectly suited for an autonomous Agentic AI solution.
    Context from Step 1:
    {business_response}
    Your task is to isolate ONE specific pain point and output your response in this precise format:
    The Core Pain Point: [The specific name of the bottleneck or operational failure point]
    Why It Breaks: [Explain why human teams or traditional rule-based software (like standard IF/THEN automation or basic RPA) struggle to solve this problem effectively]
    Financial & Operational Impact: [How this bottleneck costs the business money, leaks revenue, wastes labor hours, or introduces compliance risks]
"""

client = OpenAI(
    api_key=os.getenv("GOOGLE_API_KEY"),
    base_url=os.getenv("GEMINI_BASE_URL")
)

response = client.chat.completions.create(
    model="gemini-2.5-flash-lite",
    messages=[{"role": "user", "content": system_prompt}]
)

print(response.choices[0].message.content)
response = response.choices[0].message.content
# And repeat! In the next message, include the business idea within the message

system_prompt = f"""
{response} optimized specifically for a fintech environment. It takes the specific business area and the extracted pain point to architect the final autonomous agentic solution.
"""

response = client.chat.completions.create(
    model="gemini-2.5-flash",
    messages=[{"role": "user", "content": system_prompt}]
)
response = response.choices[0].message.content

display(Markdown(response))
```
