```python
#lets try with opensource models
from openai import OpenAI
openai = OpenAI(base_url="http://localhost:11434/v1", api_key="llama")
```

```python
def trial():

    #FIRST LLM CALL

    USER_PROMPT = "I live in India. Act as a Ventur Capitalist and an entreprenuer and " \
    "pick a business area in e-commerce domain that might be worth exploring for an Agentic AI opportunity." \
    "Remember, Agentic AI opportunity"

    response1 = openai.chat.completions.create(
        model="llama3.2",
        messages=[{
            'role':'user', 'content':USER_PROMPT
        }]
    )

    opportunity = response1.choices[0].message.content
    print(f"opportunity {opportunity}")

    #SECOND LLM CALL

    pain_point_prompt = f"""I live in India. Act as a Ventur Capitalist and an entreprenuer.You are provided with a
    business area worth exploring for an agentic ai opportunity {opportunity}. present a pain-point in that industry - something challenging that might be ripe for an Agentic solution
    and has not worked upon yet.
    """

    response2 = openai.chat.completions.create(
        model="llama3.2",
        messages=[{
            'role':'user', 'content':pain_point_prompt
        }]
    )
    idea = response2.choices[0].message.content

    print(f"Pain Point - {idea}")

    # FINAL LLM CALL

    sol_prompt = f"""
    propose a Agentic AI solution which can potentially be built and sold to businesses for the idea provided to you {idea}
    """

    response3 = openai.chat.completions.create(
        model="llama3.2",
        messages=[{
            'role':'user', 'content':sol_prompt
        }]
    )
    solution = response3.choices[0].message.content
    
    print(f"PROPOSED SOLUTION - \n {solution}")
    
```

```python
trial()
```
