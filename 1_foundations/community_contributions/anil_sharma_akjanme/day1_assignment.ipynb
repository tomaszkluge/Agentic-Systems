```python
from openai import OpenAI

client = OpenAI()
```

```python

messages = [
    {
        "role": "system",
        "content": "You are a career mentor."
    },
    {
        "role": "user",
        "content": "I want to become an AI Engineer."
    },
    {
        "role": "assistant",
        "content": "Great choice! Learn Python, ML and LLMs."
    },
    {
        "role": "user",
        "content": "Create a 6 month roadmap."
    }
]
```

```python

response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages
)

```

```python
print(response.choices[0].message.content)
```
