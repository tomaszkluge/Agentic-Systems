```python
from dotenv import load_dotenv
from agents import Agent, Runner, trace
from agents.mcp import MCPServerStdio
from IPython.display import display, Markdown

load_dotenv(override=True)
```

## Using the current date server as an MCP server

```python
instructions = "You are able to tell the current date."
request = "Which weekday is it today? Answer in a haiku in Japanese and English"
model = "gpt-4.1-mini"
```

```python
params = {"command": "uv", "args": ["run", "current_date_server.py"]}
```

```python
async with MCPServerStdio(params=params, client_session_timeout_seconds=30) as mcp_server:
    agent = Agent(name="today_manager", instructions=instructions, model=model, mcp_servers=[mcp_server])
    with trace("today_manager"):
        result = await Runner.run(agent, request)
    display(Markdown(result.final_output))
```

## Using MCP Client

```python
from current_date_client import get_current_date_tools_openai
openai_tools = await get_current_date_tools_openai()
request = "Which weekday is it today? Answer in a haiku in Japanese and English"

with trace("current_date_mcp_client"):
    agent = Agent(name="current_date_manager", instructions=instructions, model=model, tools=openai_tools)
    result = await Runner.run(agent, request)
    display(Markdown(result.final_output))
```
